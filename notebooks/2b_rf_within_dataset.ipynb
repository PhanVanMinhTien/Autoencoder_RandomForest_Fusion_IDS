```python
# Cell 1: Setup & Imports
%load_ext autoreload
%autoreload 2
import sys
import os
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split
import joblib

current_dir = Path.cwd()
root_dir = current_dir.parent
if str(root_dir) not in sys.path:
    sys.path.append(str(root_dir))

from src import config, preprocessing, rf_classifier, evaluation, utils

# Initialize experiment
exp_paths = utils.setup_experiment_folder()
print(f"Experiment ID: {exp_paths['root'].name}")

# Cell 2: Load Dataset CIC2017
X_17, y_17 = preprocessing.load_single_dataset_year('2017', binary_mode=True)

# Cell 3: Feature Selection & Conversion to NumPy
mrmr_25_features = config.mRMR_FEATURES[:25]
mrmr_indices = [config.SELECTED_FEATURES.index(f) for f in mrmr_25_features]

# Safe retrieval using .iloc and convert to NumPy array
X_17_mrmr = X_17.iloc[:, mrmr_indices].values 
y_17 = y_17.values if hasattr(y_17, 'values') else y_17

print(f"Selected 25 mRMR features. Shape: {X_17_mrmr.shape}")

# Cell 4: Splitting & Scaling
X_train, X_test, y_train, y_test = train_test_split(
    X_17_mrmr, y_17, test_size=0.2, random_state=config.SEED, stratify=y_17
)

scaler = preprocessing.get_scaler()
# Scaler learns parameters only from training set
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Save scaler
joblib.dump(scaler, exp_paths["models"] / "scaler.joblib")

# 5: Train RF
rf_save_path = exp_paths["models"] / "rf_within_2017.joblib"
rf_model = rf_classifier.train_rf(X_train_scaled, y_train, save_path=rf_save_path)

# 6: Evaluation
evaluation.evaluate_model(
    model=rf_model, 
    X_test=X_test_scaled, 
    y_test=y_test, 
    save_dir=exp_paths["figures"], 
    dataset_name="RF_Within_CIC2017"
)
```

```python
# --- CELL 2: WITHIN-DATASET CSE-CIC-IDS-2018 ---
%load_ext autoreload
%autoreload 2
import sys
import os
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split
import joblib

current_dir = Path.cwd()
root_dir = current_dir.parent
if str(root_dir) not in sys.path:
    sys.path.append(str(root_dir))

from src import config, preprocessing, rf_classifier, evaluation, utils

# Initialize experiment
exp_paths = utils.setup_experiment_folder()
print(f"Experiment ID: {exp_paths['root'].name}")
import gc

# 1. Load CSE-CIC-IDS-2018 dataset
print("Loading CSE-CIC-IDS-2018 dataset...")
X_18_raw, y_18_raw = preprocessing.load_single_dataset_year('2018', binary_mode=True)

# 2. Select features (Top 25 mRMR) and convert to NumPy
# Use same mrmr_indices list calculated in Cell 1
# Cell 3: Feature Selection & Conversion to NumPy
mrmr_25_features = config.mRMR_FEATURES[:25]
mrmr_indices = [config.SELECTED_FEATURES.index(f) for f in mrmr_25_features]

# Safe retrieval using .iloc and convert to NumPy array

X_18 = X_18_raw.iloc[:, mrmr_indices].values
y_18 = y_18_raw.values if hasattr(y_18_raw, 'values') else y_18_raw

print(f"Selected 25 mRMR features for 2018. Shape: {X_18.shape}")

# Free up RAM from raw data
del X_18_raw, y_18_raw
gc.collect()

# 3. Split data (80/20) and Standardize (Scaling)
X_train_18, X_test_18, y_train_18, y_test_18 = train_test_split(
    X_18, y_18, test_size=0.2, random_state=config.SEED, stratify=y_18
)

# Initialize new Scaler for 2018 (Avoid Data Leakage from 2017)
scaler_18 = preprocessing.get_scaler()
X_train_18_scaled = scaler_18.fit_transform(X_train_18)
X_test_18_scaled = scaler_18.transform(X_test_18)

# Save Scaler for 2018
joblib.dump(scaler_18, exp_paths["models"] / "scaler_within_2018.joblib")

# 4. Train RF model
print("Training Random Forest on CSE-CIC-IDS-2018...")
rf_model_18 = rf_classifier.train_rf(
    X_train_18_scaled, 
    y_train_18, 
    save_path=exp_paths["models"] / "rf_within_2018.joblib"
)

# 5. Evaluate results
print("\n=== EVALUATION REPORT: Within_CIC2018 ===")
evaluation.evaluate_model(
    model=rf_model_18, 
    X_test=X_test_18_scaled, 
    y_test=y_test_18, 
    save_dir=exp_paths["figures"], 
    dataset_name="RF_Within_CIC2018"
)
```
