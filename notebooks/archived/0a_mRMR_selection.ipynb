```python
# Cell 1
%load_ext autoreload
%autoreload 2

import sys
from pathlib import Path
import pandas as pd
from mrmr import mrmr_classif

# Setup path ƒë·ªÉ import src
current_dir = Path.cwd()
root_dir = current_dir.parent
if str(root_dir) not in sys.path:
    sys.path.append(str(root_dir))

from src import config, preprocessing
# Cell 2
# Ch√∫ng ta d√πng h√†m load_data_for_mrmr ƒë√£ vi·∫øt trong src/preprocessing.py
# H√†m n√†y s·∫Ω:
# 1. T·ª± ƒë·ªông t√¨m c·ªôt chung gi·ªØa 2017 v√† 2018 (kho·∫£ng 67 c·ªôt)
# 2. T·ª± ƒë·ªông Map t√™n c·ªôt 2018 -> 2017
# 3. L·∫•y m·∫´u ng·∫´u nhi√™n 200k d√≤ng (Sampling) ƒë·ªÉ ch·∫°y cho nhanh

sample_size = 200000
print(f"üöÄ ƒêang load v√† l·∫•y m·∫´u {sample_size} d√≤ng d·ªØ li·ªáu...")

# H√†m tr·∫£ v·ªÅ X_df (DataFrame ƒë·ªß c·ªôt), y (Series), v√† danh s√°ch t√™n c·ªôt
X_df, y, valid_cols = preprocessing.load_data_for_mrmr(sample_size=sample_size)

if X_df is not None:
    print(f"‚úÖ Load th√†nh c√¥ng!")
    print(f"   - K√≠ch th∆∞·ªõc d·ªØ li·ªáu: {X_df.shape}")
    print(f"   - S·ªë l∆∞·ª£ng features ·ª©ng vi√™n: {len(valid_cols)}")
    print(f"   - C√°c c·ªôt (V√≠ d·ª•): {list(X_df.columns[:5])} ...")
else:
    print("‚ùå L·ªói load d·ªØ li·ªáu. Ki·ªÉm tra l·∫°i src/config.py")
# Cell 3
if X_df is not None:
    print("‚è≥ ƒêang ch·∫°y thu·∫≠t to√°n mRMR (C√≥ th·ªÉ m·∫•t 5-10 ph√∫t)...")
    print("   (ƒêi l√†m t√°ch cafe l√† v·ª´a!)")

    # Ch·ªçn Top features
    selected_features = mrmr_classif(X=X_df, y=y, K=67)

    print("\n" + "="*50)
    print("üéâ K·∫æT QU·∫¢ TOP FEATURES")
    print("="*50)
    print(selected_features)
    print("="*50)
    print("üëâ H∆Ø·ªöNG D·∫™N: H√£y copy danh s√°ch tr√™n v√† d√°n ƒë√® v√†o bi·∫øn SELECTED_FEATURES trong file src/config.py")

```

```python
    
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src import config, preprocessing

# 1. Load l·∫°i d·ªØ li·ªáu (ch·ªâ c·∫ßn l·∫•y m·ªôt t·∫≠p m·∫´u nh·ªè ƒë·ªÉ t√≠nh t∆∞∆°ng quan cho nhanh)
print("Loading data for correlation analysis...")
df, _ = preprocessing.load_single_dataset_year('2017', binary_mode=True)
df_sample = df.sample(n=50000, random_state=config.RANDOM_STATE) # L·∫•y 50k m·∫´u l√† ƒë·ªß ƒë·ªÉ th·∫•y t∆∞∆°ng quan

# 2. X√°c ƒë·ªãnh danh s√°ch ƒë·∫∑c tr∆∞ng mRMR (Top-30 bao h√†m c·∫£ Top-25)
# L∆∞u √Ω: S·ª≠ d·ª•ng ƒë√∫ng t√™n bi·∫øn t·ª´ file config c·ªßa b·∫°n
top_30_features = config.mRMR_FEATURES[:25] 

# 3. T√≠nh to√°n ma tr·∫≠n t∆∞∆°ng quan
corr_matrix = df_sample[top_30_features].corr()

# 4. V·∫Ω bi·ªÉu ƒë·ªì Heatmap
plt.figure(figsize=(18, 14))
sns.heatmap(corr_matrix, 
            annot=False, 
            cmap='coolwarm', 
            center=0,
            linewidths=0.2)

plt.title('Correlation Heatmap of Top-30 mRMR Features', fontsize=16)
plt.tight_layout()

# 5. L∆∞u h√¨nh ·∫£nh cho b√°o c√°o
plt.savefig("mrmr_top30_correlation.png", dpi=300)
plt.show()
```
